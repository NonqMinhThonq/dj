```python
import random
import mysql.connector
from faker import Faker
from time import sleep

# Kết nối tới cơ sở dữ liệu MySQL
db = mysql.connector.connect(
    host="localhost",       # Thay bằng host của bạn
    user="root",            # Thay bằng username của bạn
    password="minhthon9",   # Thay bằng mật khẩu của bạn
    database="db2"          # Thay bằng tên database của bạn
)

cursor = db.cursor()

# Khởi tạo Faker với ngôn ngữ Việt Nam
fake = Faker("vi_VN")

# Hàm thực thi câu lệnh SQL với retry khi gặp lỗi
def execute_with_retry(cursor, sql, values, retries=3):
    for attempt in range(retries):
        try:
            cursor.executemany(sql, values)  # Chèn theo lô
            break
        except mysql.connector.errors.DatabaseError as e:
            if "Lock wait timeout exceeded" in str(e):
                print(f"Attempt {attempt + 1} failed. Retrying...")
                sleep(2)  # Chờ 2 giây trước khi thử lại
            else:
                raise e

# Fake dữ liệu và chèn vào bảng
fake_data = []
for _ in range(10):  # Tạo 10 dòng dữ liệu giả cho mỗi lần chèn
    full_name = fake.name()
    first_name = full_name.split()[-1].lower()  # Lấy tên cuối
    last_name = ''.join(full_name.split()[:-1]).lower()  # Ghép họ và tên đệm
    email = f"{first_name}{last_name}{random.randint(10, 99)}@gmail.com"
    address = fake.address()
    phone = fake.phone_number()
    academic_year = f"{random.randint(2018, 2023)}-{random.randint(2019, 2024)}"
    quote = fake.sentence(nb_words=6)
    available = random.randint(0, 1)
    status = random.randint(0, 1)
    created_by_id = 1
    major_id = random.randint(1, 10)
    university_id = 1

    # Thêm dữ liệu vào danh sách
    fake_data.append((
        full_name, email, address, phone, academic_year, quote,
        available, status, created_by_id, major_id, university_id
    ))

# Câu lệnh SQL để chèn dữ liệu
sql = """
    INSERT INTO manage_student_student 
    (name, email, address, phone, academic_year, quote, available, status, created_by_id, major_id, university_id) 
    VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)
"""

# Chia nhỏ dữ liệu và thực thi theo lô nhỏ
batch_size = 10  # Chèn 10 bản ghi một lần
for i in range(0, len(fake_data), batch_size):
    batch_data = fake_data[i:i+batch_size]
    execute_with_retry(cursor, sql, batch_data)

# Lưu thay đổi vào cơ sở dữ liệu
db.commit()

# Đóng kết nối
cursor.close()
db.close()

print("Fake data has been inserted successfully!")
```
